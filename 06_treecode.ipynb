## Step 6: Treecode

```python
import numpy
import random

N = 50                 # number of particles
n_crit = 10            # max number of particles in a single cell
theta = 0.5            # a parameter to determine far-field or near-field
epsilon = 0.01         # a small value to avoid zero division

tail = 0               # index of the last element in cells list
curr = 0               # index of the current element in the cells list

CELLS = []             # initialize the list that stores cells
x = numpy.zeros(N)
y = numpy.zeros(N)
z = numpy.zeros(N)
m = numpy.zeros(N)
phi_direct = numpy.zeros(N)   # potential array for direct summation
for i in range(N):
    x[i] = random.random()    # coordinate domain from 0 to 1
    y[i] = random.random()
    z[i] = random.random()
    m[i] = 1.0/N
```

```python
class Cell:
    def __init__(self):
        self.nleaf = 0      # number of leaves
        self.nchild = 0     # keep track of empty cells (binary counter)
        self.leaf = numpy.zeros(n_crit, dtype=numpy.int32)  # pointer to leaves
        self.x_c = self.y_c = self.z_c = 0.                # center of the cell
        self.r = 0.         # radius of the cell
        self.parent = 0     # pointer to parent cell
        self.child = numpy.zeros(8, dtype=numpy.int32)     # pointer to child cells
        self.multipole = numpy.zeros(10, dtype=numpy.float64)    # multipole array
```

```python
def add_child(octant, p, c):
    """
    ---Parameters---
    octant:  reference to one of the eight divisions in 3D
    p:       parent cell index in CELL list
    c:       child cell index in CELL list
    """
    global tail
    tail += 1
    c = tail
    CELLS.append(Cell())
    CELLS[c].r = CELLS[p].r / 2
    CELLS[c].x_c = CELLS[p].x_c + CELLS[c].r * ((octant & 1) * 2 - 1)
    CELLS[c].y_c = CELLS[p].y_c + CELLS[c].r * ((octant & 2) - 1    )
    CELLS[c].z_c = CELLS[p].z_c + CELLS[c].r * ((octant & 4) / 2 - 1)
    CELLS[c].parent = p
    CELLS[p].child[octant] = c
    CELLS[p].nchild = (CELLS[p].nchild | (1 << octant))
    print '+++cell {} is created as the child of cell {}, octant = {}'.format(c, p, octant)
```

```python
def split_cell(x, y, z, p, c):
    """
    ---Parameters---
    x, y, z: particle location arrays
    p:       parent cell index in CELL list
    c:       child cell index in CELL list
    """
    print '==start split cell {}=='.format(p)
    for i in range(n_crit):
        l = CELLS[p].leaf[i]            # l: particle index of the parent's leafs
        octant = (x[l] > CELLS[p].x_c) + ((y[l] > CELLS[p].y_c) << 1) + ((z[l] > CELLS[p].z_c) << 2)
        if not CELLS[p].nchild & (1 << octant):
            add_child(octant, p, c)
        cc = CELLS[p].child[octant]     # cc: index of the current child of parent p
        CELLS[cc].leaf[CELLS[cc].nleaf] = l
        CELLS[cc].nleaf += 1
        print '>>>particle {} is reallocated in cell {}'.format(i, cc)
        # if all the particles go into a certain child, split this child recursively
        if CELLS[cc].nleaf >= n_crit:
            split_cell(x, y, z, cc, tail)
    print '==end split cell {}=='.format(p)
```

```python
def get_multipole(p, x, y, z, m):
    if CELLS[p].nleaf >= n_crit:
        for c in range(8):
            if CELLS[p].nchild & (1 << c):
                get_multipole(CELLS[p].child[c], x, y, z, m)
    else:
        for l in range(CELLS[p].nleaf):
            j = CELLS[p].leaf[l]
            dx = CELLS[p].x_c - x[j]
            dy = CELLS[p].y_c - y[j]
            dz = CELLS[p].z_c - z[j]
            CELLS[p].multipole[0] += m[j]
            CELLS[p].multipole[1] += m[j] * dx
            CELLS[p].multipole[2] += m[j] * dy
            CELLS[p].multipole[3] += m[j] * dz
            CELLS[p].multipole[4] += m[j] * dx**2 / 2
            CELLS[p].multipole[5] += m[j] * dy**2 / 2
            CELLS[p].multipole[6] += m[j] * dz**2 / 2
            CELLS[p].multipole[7] += m[j] * dx * dy / 2
            CELLS[p].multipole[8] += m[j] * dy * dz / 2
            CELLS[p].multipole[9] += m[j] * dz * dx / 2   
```

```python
def upward_sweep(p, c):
    dx = CELLS[p].x_c - CELLS[c].x_c
    dy = CELLS[p].y_c - CELLS[c].y_c
    dz = CELLS[p].z_c - CELLS[c].z_c
    CELLS[p].multipole[0] += CELLS[c].multipole[0]
    CELLS[p].multipole[1] += CELLS[c].multipole[1] + dx * CELLS[c].multipole[0]
    CELLS[p].multipole[2] += CELLS[c].multipole[2] + dy * CELLS[c].multipole[0]
    CELLS[p].multipole[3] += CELLS[c].multipole[3] + dz * CELLS[c].multipole[0]
    CELLS[p].multipole[4] += CELLS[c].multipole[4] + dx * CELLS[c].multipole[1] + dx**2 * CELLS[c].multipole[0] / 2
    CELLS[p].multipole[5] += CELLS[c].multipole[5] + dy * CELLS[c].multipole[2] + dy**2 * CELLS[c].multipole[0] / 2
    CELLS[p].multipole[6] += CELLS[c].multipole[6] + dz * CELLS[c].multipole[3] + dz**2 * CELLS[c].multipole[0] / 2
    CELLS[p].multipole[7] += CELLS[c].multipole[7] + (dx * CELLS[c].multipole[2] + dy * CELLS[c].multipole[1] \
                                                      + dx * dy * CELLS[c].multipole[0]) / 2
    CELLS[p].multipole[8] += CELLS[c].multipole[8] + (dy * CELLS[c].multipole[3] + dz * CELLS[c].multipole[2] \
                                                      + dy * dz * CELLS[c].multipole[0]) / 2
    CELLS[p].multipole[9] += CELLS[c].multipole[9] + (dz * CELLS[c].multipole[1] + dx * CELLS[c].multipole[3] \
                                                      + dz * dx * CELLS[c].multipole[0]) / 2
```

```python
def evaluate(C, x, y, z, m, phi, i):
    global phi_tree
    if CELLS[C].nleaf >= n_crit:
        for c in range(8):
            if CELLS[C].nchild & (1 << c):
                cc = CELLS[C].child[c]
                dx = x[i] - CELLS[cc].x_c
                dy = y[i] - CELLS[cc].y_c
                dz = z[i] - CELLS[cc].z_c
                r = numpy.sqrt(dx**2 + dy**2 + dz**2)
                if CELLS[cc].r > theta*r:
                    evaluate(cc, x, y, z, m, phi_tree, i)
                else:
                    X = x[i] - CELLS[cc].x_c
                    Y = y[i] - CELLS[cc].y_c
                    Z = z[i] - CELLS[cc].z_c
                    R = numpy.sqrt(X**2 + Y**2 + Z**2)
                    R3 = R * R * R
                    R5 = R3 * R * R
                    phi_tree += CELLS[cc].multipole[0] / R
                    phi_tree += CELLS[cc].multipole[1] * (-X / R3)
                    phi_tree += CELLS[cc].multipole[2] * (-Y / R3)
                    phi_tree += CELLS[cc].multipole[3] * (-Z / R3)
                    phi_tree += CELLS[cc].multipole[4] * (3 * X * X / R5 - 1 / R3)
                    phi_tree += CELLS[cc].multipole[5] * (3 * Y * Y / R5 - 1 / R3)
                    phi_tree += CELLS[cc].multipole[6] * (3 * Z * Z / R5 - 1 / R3)
                    phi_tree += CELLS[cc].multipole[7] * (3 * X * Y / R5)
                    phi_tree += CELLS[cc].multipole[8] * (3 * Y * Z / R5)
                    phi_tree += CELLS[cc].multipole[9] * (3 * Z * X / R5)
    else:
        for l in range(CELLS[C].nleaf):
            j = CELLS[C].leaf[l]
            dx = x[i] - x[j]
            dy = y[i] - y[j]
            dz = z[i] - z[j]
            r = numpy.sqrt(dx**2 + dy**2 + dz**2 + epsilon**2)
            phi_tree += m[j] / r
```

```python
# direct summation
for i in range(N):
    phi_direct[i] = -m[i]/epsilon    # offset the additional term when i = j
    for j in range(N):
        dx = x[i] - x[j]
        dy = y[i] - y[j]
        dz = z[i] - z[j]
        r = numpy.sqrt(dx**2 + dy**2 + dz**2 + epsilon**2)
        phi_direct[i] += m[j]/r

# set root cell
CELLS.append(Cell())
CELLS[0].x_c = CELLS[0].y_c = CELLS[0].z_c = 0.5
CELLS[0].r = 0.5

# build tree
for i in range(N):
    curr = 0     # self counter (self)
    while CELLS[curr].nleaf >= n_crit:
        CELLS[curr].nleaf += 1
        octant = (x[i] > CELLS[curr].x_c) + ((y[i] > CELLS[curr].y_c) << 1) + ((z[i] > CELLS[curr].z_c) << 2)
        if not CELLS[curr].nchild & (1 << octant):
            add_child(octant, curr, tail)
        curr = CELLS[curr].child[octant]
    CELLS[curr].leaf[CELLS[curr].nleaf] = i
    CELLS[curr].nleaf += 1
    print 'particle {} is stored in cell {}'.format(i, curr)
    if CELLS[curr].nleaf >= n_crit:
        split_cell(x, y, z, curr, tail)

# multipole expansion
get_multipole(0, x, y, z, m)

# upward translation
for c in range(tail, 0, -1):
    p = CELLS[c].parent
    upward_sweep(p, c)

# evaluate expansion
err = 0.
rel = 0.
for i in range(N):
    phi_tree = -m[i]/epsilon
    evaluate(0, x, y, z, m, phi_tree, i)
    err += (phi_direct[i] - phi_tree)**2
    rel += phi_direct[i]**2
    print '{} {} {}'.format(i, phi_direct[i], phi_tree)
print 'error : {}'.format(numpy.sqrt(err/rel))
```
