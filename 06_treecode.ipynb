##### Potential Evaluation

Based on the tree structure, we distribute each particle in a certain twig cell, so that when we consider the influence of source points, we no longer need to loop in each source particle, instead we should loop in each cell. The best way to go over every cell is to do a traverse from the root cell. Considering we are traversing the tree to calculate the potential of a certain target particle $i$: if the current cell is in 'near' to particle $i$, we will use direct summation to calculate this near-field contribution; while if the current cell is relatively 'far' to particle $i$, then we evaluate the far-field effect using multipole expansion. So the next question is how can we distinguish between 'near' and 'far' ?

Barnet and Hut firstly introduced this criterion in their paper published in 1986:

- near field: $$\frac{l_{cell}}{d} > \theta $$ 

- far field: $$\frac{l_{cell}}{d} \leq \theta $$

$r_{cell}$ denotes the length of current cell, $d$ is the distance between the target point and the center of the cell, and $\theta$ is the tolerance parameter, a fixed value around 1. Recall in our cell data structure, we use radius of a cell rather than length ($l_{cell} = 2r_{cell}$). Thus, the criterion applied in our treecode can be written as:

- near field: $$ r_{cell} > \theta \cdot d$$

- far field: $$ r_{cell} \leq \theta \cdot d$$

where $\theta$ is around 0.5.

<img src="image/potential_eval.png">

The figure above shows the flowchart of potential evaluation module. For each target particle $i$, we traverse from the root cell. If the current cell is a twig, then we use direct summation to add the cell's contribution to the potential. If not, we use the criterion metioned above to identify whether the current cell is in near-field or far-field. For a far-field non-twig cell, we apply multipole expansion to accumulate into the potential. For a near-field non-twig cell, we need to recursively traverse down to twig cells, in order to go over each source particles located in these near-field twig cells to perform direct summation.

```python
def evaluate(phi, x, y, z, m, p, i, cells, n_crit, theta, epsilon):
    """Evaluate the gravitational potential at target point i, caused by source particles cell p. If leaf number of cell p is less than n_crit (twig), use direct summation. Otherwise (non-twig), loop in p's child cells. If child cell c is in far-field of target particle i, use multipole expansion. Otherwise (near-field), call the function recursively.
    
    Arguments:
        phi:     potential of target particle i
        x, y, z: coordinate arrays of particles
        m:       mass array of particles
        p:       cell index in cells list
        i:       target particle index
        cells:   the list of cells
        n_crit:  maximum number of leaves in a single cell
        theta:   tolerance parameter
        epsilon: a small value to avoid zero division
    
    Returns:
        phi:     potential of target particle i
    """
    
    eps2 = epsilon**2
    # non-twig cell
    if cells[p].nleaf >= n_crit:
        # loop in p's child cells (8 octants)
        for c in range(8):
            if cells[p].nchild & (1 << c):
                c = cells[p].child[c]
                dx = x[i] - cells[c].x_c
                dy = y[i] - cells[c].y_c
                dz = z[i] - cells[c].z_c
                r = numpy.sqrt(dx**2 + dy**2 + dz**2)
                # near-field child cell
                if cells[c].r > theta*r:
                    phi = evaluate(phi, x, y, z, m, c, i, cells, n_crit, theta, epsilon)
                # far-field child cell
                else:
                    X = x[i] - cells[c].x_c
                    Y = y[i] - cells[c].y_c
                    Z = z[i] - cells[c].z_c
                    R = numpy.sqrt(X**2 + Y**2 + Z**2)
                    R3 = R * R * R
                    R5 = R3 * R * R
                    phi += cells[c].multipole[0] / R
                    phi += cells[c].multipole[1] * (-X / R3)
                    phi += cells[c].multipole[2] * (-Y / R3)
                    phi += cells[c].multipole[3] * (-Z / R3)
                    phi += cells[c].multipole[4] * (3 * X * X / R5 - 1 / R3)
                    phi += cells[c].multipole[5] * (3 * Y * Y / R5 - 1 / R3)
                    phi += cells[c].multipole[6] * (3 * Z * Z / R5 - 1 / R3)
                    phi += cells[c].multipole[7] * (3 * X * Y / R5)
                    phi += cells[c].multipole[8] * (3 * Y * Z / R5)
                    phi += cells[c].multipole[9] * (3 * Z * X / R5)
    #twig cell
    else:
        # loop in twig cell's particles
        for l in range(cells[p].nleaf):
            j = cells[p].leaf[l]
            dx = x[i] - x[j]
            dy = y[i] - y[j]
            dz = z[i] - z[j]
            r = numpy.sqrt(dx**2 + dy**2 + dz**2 + eps2)
            phi += m[j] / r
    return phi
```

The above section is the last module in the treecode. We introduce the criterion to compare $r_{cell}$ with $\theta \cdot r$. A bigger $\theta$ implies a less strict requirement of "far-field", therefore, a bigger $\theta$ will bring about a greater calculation error. On the contrary, a smaller $\theta$ indicates a more accurate solution. 

##### Final Treecode

In notebook 1, we talked about the direct summation, which is our knowledge before this tutorial. Then we elaborated how to formulate multipole expansion in notebook 2 & 3, and call it particle to multipole (P2M). In notebook 4, we introduced the cell class, and based on which we generate the tree structure using a list. In the previous one, we covered the upward sweep, by which a parent cell's multipole can be evaluated by traversing over its child cells, and that is why we name it as multipole to multipole (M2M). We just went over the potential evaluation module. Now it is time to integrate these modules into our final treecode. Each function we defined before has been added to the python script (treecode.py).

Before the code, we redeclare the problem statement and the assumptions: there are plenty of particles randomly scattering in domain $x$, $y$, $z$ $\in$ $\left[ 0, 1 \right]$, we assume they have the same mass and the total mass is equal to 1.

```python
import numpy
import random
import time
import treecode as tree
import matplotlib.pyplot as pyplot
%matplotlib inline

n = 1000           # number of particles
n_crit = 10      # max number of particles in a single cell
theta = 0.5      # a parameter to determine far-field or near-field
epsilon = 0.01   # a small value to avoid zero division

root = 0         # index of the first element in cells list
tail = 0         # index of the last element in cells list
curr = 0         # index of the current element in the cells list

cells = []                    # initialize the cells list
x = numpy.zeros(n)            # initialize the coordinates array
y = numpy.zeros(n)
z = numpy.zeros(n)
m = numpy.zeros(n)            # initialize the mass array
phi_direct = numpy.zeros(n)   # initialize the potential array for direct summation
phi_tree = numpy.zeros(n)     # initialize the potential array for treecode

for i in range(n):
    x[i] = random.random()    # generate coordinates from 0 to 1
    y[i] = random.random()
    z[i] = random.random()
    m[i] = 1.0/n              # assume the same mass
```

We called time.clock() function before and after each module in order to study the breakdown of computation time.

```python
# direct summation
tic = time.clock()
phi_direct = tree.direct_sum(n, x, y, z, m, epsilon)
toc = time.clock()
time_direct = toc - tic

# set root cell
tic = time.clock()
cells.append(tree.Cell(n_crit))
cells[0].x_c = cells[0].y_c = cells[0].z_c = 0.5
cells[0].r = 0.5

# build tree
for i in range(n):
    curr = 0
    while cells[curr].nleaf >= n_crit:
        cells[curr].nleaf += 1
        octant = (x[i] > cells[curr].x_c) + ((y[i] > cells[curr].y_c) << 1) \
               + ((z[i] > cells[curr].z_c) << 2)
        
        if not cells[curr].nchild & (1 << octant):
            tree.add_child(octant, curr, cells, n_crit)
        curr = cells[curr].child[octant]
    cells[curr].leaf[cells[curr].nleaf] = i
    cells[curr].nleaf += 1
    
    if cells[curr].nleaf >= n_crit:
        tree.split_cell(x, y, z, curr, cells, n_crit)

toc = time.clock()
time_src = toc - tic
        
# multipole expansion (P2M: Particle to Multipole)
tic = time.clock()
tree.get_multipole(x, y, z, m, root, cells, n_crit)
toc = time.clock()
time_P2M = toc - tic

# upward translation (M2M: multipole to multipole)
tic = time.clock()
tail = len(cells) - 1
for c in range(tail, root, -1):
    p = cells[c].parent
    tree.upward_sweep(p, c, cells)

toc = time.clock()
time_M2M = toc - tic

# evaluate force
tic = time.clock()
phi_tree = - m/epsilon
for i in range(n):
    phi_tree[i] = evaluate(phi_tree[i], x, y, z, m, root, i, cells, n_crit, theta, epsilon)

toc = time.clock()
time_eval = toc - tic
```

##### Error Analysis

```python
# relative error
err_rel = abs((phi_tree - phi_direct)/phi_direct)
# relative error in l2 norm (Euclidean norm) 
err_l2 = numpy.linalg.norm(err_rel, 2)
print err_l2
```

We calculate the relative error for each target particle, and use $L^2$ norm (Euclidean norm) to evaluate the overall error. Based on the error value, we generate the following scatter plots. The upper plot demonstrates the potential distribution of direct summation method and treecode. It is intuitive to see whether the result of direct summation and treecode are matching well to each other, and the lower one shows the magnitude of the relative error for each particle's potential.

```python
from matplotlib import rcParams
rcParams['figure.dpi'] = 100
rcParams['font.size'] = 16
rcParams['font.family'] = 'StixGeneral'

pyplot.figure(figsize=(12,8))
pyplot.subplot(2, 1, 1)
f1 = pyplot.scatter(range(1, n+1), phi_direct, s=30, marker='o', c='r', alpha=0.7)
f2 = pyplot.scatter(range(1, n+1), phi_tree, s=30, marker='x', c='b', alpha=0.8)
pyplot.xlim(0,n)
pyplot.ylabel('$\phi_{direct} , \phi_{tree}$')
pyplot.legend((f1, f2), ('Direct', 'Tree'), scatterpoints=2, loc='upper left', prop={'size':12})

ax = pyplot.subplot(2, 1, 2)
pyplot.scatter(range(1, n+1), err_rel, s=30, c='g', alpha=0.6)
pyplot.xlim(0,n)
pyplot.ylim(1e-6, 1e-1)
pyplot.xlabel('Particle Index')
pyplot.ylabel('$e_{rel}$')
ax.set_yscale('log')

pyplot.show();
```

Recall the tolerance parameter $\theta$, which strongly affects the level of treecode accuracy. A larger $\theta$ means less strict definition of "far-field", thus giving rise to a greater error. To verify this trend, here we perform a parameter study by giving a set of $\theta$ decreasing from 0.5 to 0.1. (particle number $n$ = 1000, critical number $n_{crit}$  = 10)

```python
theta_set = numpy.linspace(0.5, 0.1, 5)
err_set = [0.0607390697452, 0.03363727980938, 0.0153864740105, 0.006673603354102, 0.00124844511301]
pyplot.figure(figsize=(6,4))
pyplot.plot(theta_set, err_set, linewidth=2)
pyplot.xlabel(r'$\theta$')
pyplot.ylabel(r'$e_{rel}$ in $L^2$ norm');
```

The figure above illustrates that the $L^2$ norm error increases as the tolerance parameter $\theta$ goes up. Note that when $\theta$ approaches to 0, no particle is considered in the far-field, therefore, the treecode is drawing near to direct summation.
