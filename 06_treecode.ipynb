## Step 6: Treecode

```python
import numpy
import random

n = 50           # number of particles
n_crit = 10      # max number of particles in a single cell
theta = 0.5      # a parameter to determine far-field or near-field
epsilon = 0.01   # a small value to avoid zero division

root = 0         # index of the first element in cells list
tail = 0         # index of the last element in cells list
curr = 0         # index of the current element in the cells list

cells = []                    # initialize the cells list
x = numpy.zeros(n)            # initialize the coordinates array
y = numpy.zeros(n)
z = numpy.zeros(n)
m = numpy.zeros(n)            # initialize the mass array
phi_direct = numpy.zeros(n)   # initialize the potential array for direct summation
for i in range(n):
    x[i] = random.random()    # generate coordinates from 0 to 1
    y[i] = random.random()
    z[i] = random.random()
    m[i] = 1.0/n              # assume the same mass
```

```python
class Cell:
    """The class for a cell"""
    def __init__(self):
        self.nleaf = 0        # number of leaves
        self.nchild = 0       # binary counter to keep track of empty cells
        self.leaf = numpy.zeros(n_crit, dtype=numpy.int32)     # array of leaf index
        self.x_c = self.y_c = self.z_c = 0.                    # center of the cell
        self.r = 0.           # radius of the cell
        self.parent = 0       # index of parent cell
        self.child = numpy.zeros(8, dtype=numpy.int32)         # array of child index
        self.multipole = numpy.zeros(10, dtype=numpy.float64)  # multipole array
```

```python
def add_child(octant, p):
    """Add a cell to the end of cells list as a child of p, initialize the center and
       radius of the child cell c, and establish mutual reference between child c and
       parent p
    
    Arguments:
      octant:  reference to one of the eight divisions in three dimensions
      p:       parent cell index in cells list
    """
    global tail
    tail += 1
    c = tail
    cells.append(Cell())
    cells[c].r = cells[p].r / 2
    cells[c].x_c = cells[p].x_c + cells[c].r * ((octant & 1) * 2 - 1)
    cells[c].y_c = cells[p].y_c + cells[c].r * ((octant & 2) - 1    )
    cells[c].z_c = cells[p].z_c + cells[c].r * ((octant & 4) / 2 - 1)
    cells[c].parent = p
    cells[p].child[octant] = c
    cells[p].nchild = (cells[p].nchild | (1 << octant))
```

```python
def split_cell(p):
    """Loop in parent p's leafs and reallocate the particles to subcells. If a subcell
       has not been created in that octant, create one using add_child. If the subcell
       c's leaf number exceeds n_crit, split the subcell c recursively.
    
    Arguments: 
    p:       parent cell index in cells list
    """
    for i in range(n_crit):
        l = cells[p].leaf[i]
        octant = (x[l] > cells[p].x_c) + ((y[l] > cells[p].y_c) << 1) \
               + ((z[l] > cells[p].z_c) << 2)
        
        if not cells[p].nchild & (1 << octant):
            add_child(octant, p)
        
        c = cells[p].child[octant]
        cells[c].leaf[cells[c].nleaf] = l
        cells[c].nleaf += 1
        
        if cells[c].nleaf >= n_crit:
            split_cell(c)
```

```python
def get_multipole(p):
    """Calculate multipole arrays for all twig cells under cell p. If leaf number of 
       cell p is bigger than n_crit (non-twig), traverse down recursively. Otherwise
       (twig), calculate the multipole arrays for twig cell p.
    
    Arguments:
    p:       cell index in cells list
    """
    if cells[p].nleaf >= n_crit:
        for c in range(8):
            if cells[p].nchild & (1 << c):
                get_multipole(cells[p].child[c])
    else:
        for l in range(cells[p].nleaf):
            j = cells[p].leaf[l]
            dx = cells[p].x_c - x[j]
            dy = cells[p].y_c - y[j]
            dz = cells[p].z_c - z[j]
            cells[p].multipole[0] += m[j]
            cells[p].multipole[1] += m[j] * dx
            cells[p].multipole[2] += m[j] * dy
            cells[p].multipole[3] += m[j] * dz
            cells[p].multipole[4] += m[j] * dx * dx / 2
            cells[p].multipole[5] += m[j] * dy * dy / 2
            cells[p].multipole[6] += m[j] * dz * dz / 2
            cells[p].multipole[7] += m[j] * dx * dy / 2
            cells[p].multipole[8] += m[j] * dy * dz / 2
            cells[p].multipole[9] += m[j] * dz * dx / 2   
```

```python
def upward_sweep(p, c):
    """Calculate parent cell p's multipole array based on child cell c's multipoles
    
    Arguments:
    p:       parent cell index in cells list
    c:       child cell index in cells list
    """
    dx = cells[p].x_c - cells[c].x_c
    dy = cells[p].y_c - cells[c].y_c
    dz = cells[p].z_c - cells[c].z_c
    cells[p].multipole[0] += cells[c].multipole[0]
    cells[p].multipole[1] += cells[c].multipole[1] + dx * cells[c].multipole[0]
    cells[p].multipole[2] += cells[c].multipole[2] + dy * cells[c].multipole[0]
    cells[p].multipole[3] += cells[c].multipole[3] + dz * cells[c].multipole[0]
    cells[p].multipole[4] += cells[c].multipole[4] + dx * cells[c].multipole[1] \
                                                   + dx * dx * cells[c].multipole[0] / 2
    cells[p].multipole[5] += cells[c].multipole[5] + dy * cells[c].multipole[2] \
                                                   + dy * dy * cells[c].multipole[0] / 2
    cells[p].multipole[6] += cells[c].multipole[6] + dz * cells[c].multipole[3] \
                                                   + dz * dz * cells[c].multipole[0] / 2
    cells[p].multipole[7] += cells[c].multipole[7] + (dx * cells[c].multipole[2] \
                                                   +  dy * cells[c].multipole[1] \
                                                   +  dx * dy * cells[c].multipole[0]) / 2
    cells[p].multipole[8] += cells[c].multipole[8] + (dy * cells[c].multipole[3] \
                                                   +  dz * cells[c].multipole[2] \
                                                   +  dy * dz * cells[c].multipole[0]) / 2
    cells[p].multipole[9] += cells[c].multipole[9] + (dz * cells[c].multipole[1] \
                                                   +  dx * cells[c].multipole[3] \
                                                   +  dz * dx * cells[c].multipole[0]) / 2
```

```python
def evaluate(p, i):
    """Evaluate the gravitational potential at target point i, caused by source particles
       cell p. If leaf number of cell p is less than n_crit (twig), use direct summation.
       Otherwise (non-twig), loop in p's child cells. If child cell c is in far-field of
       target particle i, use multipole expansion.Otherwise (near-field), call the 
       function recursively.
    Arguments:
    p:       cell index in cells list
    i:       target particle index
    """
    global phi_tree
    if cells[p].nleaf >= n_crit:
        for c in range(8):
            if cells[p].nchild & (1 << c):
                c = cells[p].child[c]
                dx = x[i] - cells[c].x_c
                dy = y[i] - cells[c].y_c
                dz = z[i] - cells[c].z_c
                r = numpy.sqrt(dx**2 + dy**2 + dz**2)
                if cells[c].r > theta*r:
                    evaluate(c, i)
                else:
                    X = x[i] - cells[c].x_c
                    Y = y[i] - cells[c].y_c
                    Z = z[i] - cells[c].z_c
                    R = numpy.sqrt(X**2 + Y**2 + Z**2)
                    R3 = R * R * R
                    R5 = R3 * R * R
                    phi_tree += cells[c].multipole[0] / R
                    phi_tree += cells[c].multipole[1] * (-X / R3)
                    phi_tree += cells[c].multipole[2] * (-Y / R3)
                    phi_tree += cells[c].multipole[3] * (-Z / R3)
                    phi_tree += cells[c].multipole[4] * (3 * X * X / R5 - 1 / R3)
                    phi_tree += cells[c].multipole[5] * (3 * Y * Y / R5 - 1 / R3)
                    phi_tree += cells[c].multipole[6] * (3 * Z * Z / R5 - 1 / R3)
                    phi_tree += cells[c].multipole[7] * (3 * X * Y / R5)
                    phi_tree += cells[c].multipole[8] * (3 * Y * Z / R5)
                    phi_tree += cells[c].multipole[9] * (3 * Z * X / R5)
    else:
        for l in range(cells[p].nleaf):
            j = cells[p].leaf[l]
            dx = x[i] - x[j]
            dy = y[i] - y[j]
            dz = z[i] - z[j]
            r = numpy.sqrt(dx**2 + dy**2 + dz**2 + epsilon**2)
            phi_tree += m[j] / r
```

```python
# direct summation
for i in range(n):
    phi_direct[i] = -m[i]/epsilon
    for j in range(n):
        dx = x[i] - x[j]
        dy = y[i] - y[j]
        dz = z[i] - z[j]
        r = numpy.sqrt(dx**2 + dy**2 + dz**2 + epsilon**2)
        phi_direct[i] += m[j]/r

# set root cell
cells.append(Cell())
cells[0].x_c = cells[0].y_c = cells[0].z_c = 0.5
cells[0].r = 0.5

# build tree
for i in range(n):
    curr = 0     # self counter (self)
    while cells[curr].nleaf >= n_crit:
        cells[curr].nleaf += 1
        octant = (x[i] > cells[curr].x_c) + ((y[i] > cells[curr].y_c) << 1) \
               + ((z[i] > cells[curr].z_c) << 2)
        
        if not cells[curr].nchild & (1 << octant):
            add_child(octant, curr)
        curr = cells[curr].child[octant]
    cells[curr].leaf[cells[curr].nleaf] = i
    cells[curr].nleaf += 1
    
    if cells[curr].nleaf >= n_crit:
        split_cell(curr)

# multipole expansion
get_multipole(root)

# upward translation
for c in range(tail, root, -1):
    p = cells[c].parent
    upward_sweep(p, c)

# evaluate expansion
err = 0.
rel = 0.
print 'id   phi_direct   phi_tree'
for i in range(n):
    phi_tree = -m[i]/epsilon
    evaluate(root, i)
    err += (phi_direct[i] - phi_tree)**2
    rel += phi_direct[i]**2
    print '{}  {} {}'.format(i, phi_direct[i], phi_tree)
print 'error : {}'.format(numpy.sqrt(err/rel))
```
